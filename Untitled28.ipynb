```python
# Task 2: Analyze Characteristics of Duplicates
# Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
# Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Define the subset of columns to check for duplicates
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']

# 1. Find duplicates based on the selected columns
duplicates_subset = df[df.duplicated(subset=subset_cols, keep=False)]

# 2. Count the number of duplicates
num_duplicates_subset = len(duplicates_subset)
print(f"Number of duplicate rows (based on {subset_cols}): {num_duplicates_subset}")

# 3. Display the first few duplicates for inspection
if num_duplicates_subset > 0:
    print("\nFirst few duplicate rows (based on selected columns):")
    print(duplicates_subset.sort_values(by=subset_cols).head())
else:
    print("\nNo duplicates found based on the selected columns.")

# 4. Analyze which other columns frequently match in duplicates
if num_duplicates_subset > 0:
    print("\nFrequency of identical values in other columns among duplicates:")
    for column in df.columns:
        if column not in subset_cols:
            # Check if the column has identical values in each duplicate group
            grouped_duplicates = duplicates_subset.groupby(subset_cols)[column]
            identical_percentage = (grouped_duplicates.nunique() == 1).mean() * 100
            print(f"{column}: {identical_percentage:.2f}% identical in duplicate groups")
else:
    print("\nNo duplicates to analyze further.")
```

```python
#  Load Data & Identify Duplicates
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Define columns to check for duplicates
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']
duplicates = df[df.duplicated(subset=subset_cols, keep=False)]
print(f"Total duplicate rows (based on {subset_cols}): {len(duplicates)}")
```

```python
# Bar Chart: Duplicates by Country (Top 10)
```

```python
# Count duplicates per country (Top 10)
dup_by_country = duplicates['Country'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=dup_by_country.values, y=dup_by_country.index, palette="viridis")
plt.title("Top 10 Countries with Most Duplicates")
plt.xlabel("Number of Duplicates")
plt.ylabel("Country")
plt.show()
```

```python
# Pie Chart: Duplicates by Employment Type
```

```python
# Count duplicates per employment type
dup_by_employment = duplicates['Employment'].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(
    dup_by_employment,
    labels=dup_by_employment.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("pastel"),
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Distribution of Duplicates by Employment Type")
plt.show()
```

```python
# Heatmap: Country vs. Employment (Top 10 Countries)
```

```python
# Cross-tabulation of Country vs. Employment (Top 10 countries)
top_countries = duplicates['Country'].value_counts().head(10).index
dup_filtered = duplicates[duplicates['Country'].isin(top_countries)]
dup_cross = pd.crosstab(dup_filtered['Country'], dup_filtered['Employment'])

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(dup_cross, annot=True, cmap="Blues", fmt="d", linewidths=0.5)
plt.title("Duplicates: Country vs. Employment (Top 10 Countries)")
plt.show()
```
